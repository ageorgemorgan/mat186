This script creates a plot used in a great problem devised by Alfonso Gracia-Saz. 

```python
import numpy as np

import matplotlib.pyplot as plt

import cmocean
import cmocean.cm as cmo

import time

import sys

# this makes sure the images don't come out looking muddy on 
# retina displays (which your computer probably uses)
%config InlineBackend.figure_format='retina'
```

```python
def f(xx):
    f = np.zeros_like(xx)
    for j in np.arange(0, np.shape(f)[0]):
        x = xx[j]
        funclist = [lambda x: 2., lambda x : (x+1)**2+1, lambda x: 2-2*x, lambda x: 2-2*x, lambda x: 6-2*x]
        f[j] = np.piecewise(x, [x<=-2, -2<x<0, 0<= x<1, 1<x<2, x>2], funclist)
    return f
```

```python
fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.rcParams['figure.dpi']= 600

ax.grid()
ax.set_axisbelow(True)

x = np.linspace(-4,0.97, 1000)
y = f(x)
plt.plot(x,y, color='xkcd:barney purple', linewidth='2', linestyle='solid')
x = np.linspace(1.03,1.97, 1000)
y = f(x)
plt.plot(x,y, color='xkcd:barney purple', linewidth='2', linestyle='solid')
x = np.linspace(2.03,4, 1000)
y = f(x)
plt.plot(x,y, color='xkcd:barney purple', linewidth='2', linestyle='solid', label='$f(x)$')

plt.scatter([1,2,2], [0,2,-2], facecolors='w', edgecolors='xkcd:barney purple', s=70)
plt.scatter([2,4], [-1,-2], facecolors='xkcd:barney purple', edgecolors='xkcd:barney purple', s=70)

plt.xlim([-4,4.1])
plt.ylim([-2.2,2.2])

textcolor = 'k'

plt.xlabel(r"$x$", fontsize=26, color = textcolor)
plt.ylabel(r"$y$", fontsize=26, color=textcolor)

plt.tick_params(axis='x', which='both', top='off', color = textcolor)
plt.xticks(fontsize=24, rotation=0, color = textcolor)
plt.tick_params(axis='y', which='both', right='off', color = textcolor)
plt.yticks(fontsize=24, rotation=0, color = textcolor)
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=12)

ax.legend(fontsize=24, loc='center left')

plt.tight_layout()

fig.set_size_inches(8, 6)

plt.savefig('alfonsos_famous_function'+'.jpg', dpi=600)
plt.show()
```

Alfonso's problem is: compute 
$$
\lim_{x\rightarrow 0} f(f(x)). 
$$
