This is code for plotting functions of the form 
$$
f_{n}(x) = x^{n} \sin\left(\frac{1}{x}\right). 
$$
Following Spivak's calculus book, these functions are very useful when it comes to teaching limits and the "baby ideas" of $\delta-\epsilon$ proofs. 

The labelling/saving script is a bit sloppy and more stuff should be automated so that changing $n$ is easy, but whatever. Also I should probably be using a Chebyshev grid for drawing instead of a uniform one (or, if I really want a unif. one, I should use splines) but I've spent enough time on this already. Next time I teach this stuff I will clean it up and make it really nice. 

```python
import numpy as np

import matplotlib.pyplot as plt

import cmocean
import cmocean.cm as cmo

import time

import sys

# this makes sure the images don't come out looking muddy on 
# retina displays (which your computer probably uses)
%config InlineBackend.figure_format='retina'
```

```python
fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.rcParams['figure.dpi']= 600
    
delta = 1e-4
x = np.linspace(-0.1,-delta, 10000)
xx = np.linspace(delta, 0.1,10000)
n = 0.
y = (x**n)*np.sin(1./x)
yy = (xx**n)*np.sin(1./xx)

plt.plot(x, y, color='xkcd:medium pink', linewidth='2', linestyle='solid')
plt.plot(xx, yy, color='xkcd:medium pink', linewidth='2', linestyle='solid', label='$f_{0}(x)$')

"""
plt.plot(x, x**n, color='xkcd:nice blue', linewidth='3', linestyle='dashed')
plt.plot(xx, xx**n, color='xkcd:nice blue', linewidth='3', linestyle='dashed')

plt.plot(x, -(x**n), color='xkcd:nice blue', linewidth='3', linestyle='dashed')
plt.plot(xx, -(xx**n), color='xkcd:nice blue', linewidth='3', linestyle='dashed', label='$\pm1$')
"""

ax.legend(fontsize=16)

plt.xlim([-0.1,0.1])
plt.ylim([-1.01,1.01])

textcolor = 'k'

plt.xlabel(r"$x$", fontsize=26, color = textcolor)
plt.ylabel(r"$y$", fontsize=26, color=textcolor)

plt.tick_params(axis='x', which='both', top='off', color = textcolor)
plt.xticks(fontsize=24, rotation=0, color = textcolor)
plt.tick_params(axis='y', which='both', right='off', color = textcolor)
plt.yticks(fontsize=24, rotation=0, color = textcolor)
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=4)

plt.tight_layout()

fig.set_size_inches(8, 6)

#ax.axes.set_aspect(scaled)
plt.savefig('wobbly_boy_newest'+'.jpg', dpi=600)
plt.show()
```
