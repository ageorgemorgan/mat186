```python
import numpy as np

import matplotlib.pyplot as plt

import cmocean
import cmocean.cm as cmo

import time

import sys

# this makes sure the images don't come out looking muddy on 
# retina displays (which your computer probably uses)
%config InlineBackend.figure_format='retina'
```

```python
def g(xx):
    g = np.zeros_like(xx)
    for j in np.arange(0, np.shape(g)[0]):
        x = xx[j]
        funclist = [lambda x: 3., lambda x : 1.+np.abs(x), lambda x : 1.+2*np.abs(x), lambda x: 1., lambda x: 0.]
        g[j] = np.piecewise(x, [x<=-1, -1<x<=0, 0<x<1,x==1, x>1], funclist)
    return g

def f(xx):
    f = np.zeros_like(xx)
    for j in np.arange(0, np.shape(f)[0]):
        x = xx[j]
        funclist = [lambda x: 0, lambda x : 1.]
        f[j] = np.piecewise(x, [x<=0, x>0], funclist)
    return f
```

```python
fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.rcParams['figure.dpi']= 600

ax.grid()
ax.set_axisbelow(True)

plt.scatter([-1,1,1], [3,3,0], facecolors='w', edgecolors='xkcd:carnation', s=70)
plt.scatter([-1,1], [2,2], facecolors='xkcd:carnation', edgecolors='xkcd:carnation', s=70)
plt.scatter([0], [1], facecolors='w', edgecolors='xkcd:jade green', s=70)
plt.scatter([0], [0], facecolors='xkcd:jade green', edgecolors='xkcd:jade green', s=70)

x = np.linspace(-3,-1.06, 100)
y = g(x)
plt.plot(x,y, color='xkcd:carnation', linewidth='2.5', linestyle='solid')
x = np.linspace(-0.96,0.98, 1000)
y = g(x)
plt.plot(x,y, color='xkcd:carnation', linewidth='2.5', linestyle='solid')
x = np.linspace(1.065,3, 100)
y = g(x)
plt.plot(x,y, color='xkcd:carnation', linewidth='2.5', linestyle='solid', label='$g(x)$')

x = np.linspace(-3,-0.01, 1000)
y = f(x)
plt.plot(x,y, color='xkcd:jade green', linewidth='2.5', linestyle='dashed')
x = np.linspace(.045,3, 1000)
y = f(x)
plt.plot(x,y, color='xkcd:jade green', linewidth='2.5', linestyle='dashed', label='$f(x)$')


plt.xlim([-3.,3.])
#plt.ylim([-2.2,2.2])

textcolor = 'k'

plt.xlabel(r"$x$", fontsize=26, color = textcolor)
plt.ylabel(r"$y$", fontsize=26, color=textcolor)

plt.tick_params(axis='x', which='both', top='off', color = textcolor)
plt.xticks(fontsize=24, rotation=0, color = textcolor)
plt.tick_params(axis='y', which='both', right='off', color = textcolor)
plt.yticks(fontsize=24, rotation=0, color = textcolor)
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=6)

ax.legend(fontsize=24, loc='center left')

plt.tight_layout()

fig.set_size_inches(8, 6)

plt.savefig('funny_compositions'+'.jpg', dpi=600)
plt.show()
```
