```python
import numpy as np

import matplotlib.pyplot as plt

import time

import sys

# this makes sure the images don't come out looking muddy on 
# retina displays (which your computer probably uses)
%config InlineBackend.figure_format='retina'
```

```python
fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.rcParams['figure.dpi']= 600

ax.grid()
ax.set_axisbelow(True)

plt.scatter([-1], [-1], facecolors='xkcd:jade green', edgecolors='xkcd:jade green', s=70)

x = np.linspace(-3,3, 600)
y = x**3
plt.plot(x,y, color='xkcd:jade green', linewidth='2.5', linestyle='solid', label='$f(x)=x^3$')

yy = 3*(x+1)-1
plt.plot(x,yy, color='xkcd:carnation', linewidth='2.5', linestyle='dashed', label='Tangent to $f(x)$ at $x=-1$')

plt.xlim([-2.2,2.5])
plt.ylim([-12, 16])

textcolor = 'k'

plt.xlabel(r"$x$", fontsize=26, color = textcolor)
plt.ylabel(r"$y$", fontsize=26, color=textcolor)

plt.tick_params(axis='x', which='both', top='off', color = textcolor)
plt.xticks(fontsize=24, rotation=0, color = textcolor)
plt.tick_params(axis='y', which='both', right='off', color = textcolor)
plt.yticks(fontsize=24, rotation=0, color = textcolor)
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=6)

ax.legend(fontsize=20, loc='upper left')

plt.tight_layout()

fig.set_size_inches(8, 6)

plt.savefig('cubic_tangents'+'.jpg', dpi=600)
plt.show()
```

^ illustrates that a tangent lying above/below a graph is a LOCAL definition. 

```python
fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.rcParams['figure.dpi']= 600

ax.grid()
ax.set_axisbelow(True)

plt.scatter([0,1], [1,0], facecolors='xkcd:raspberry', edgecolors='xkcd:raspberry', s=70)

t = np.linspace(0,1, 600)
x = ((t-1)**2)**(2./3.)
plt.plot(t,x, color='xkcd:raspberry', linewidth='2.5', linestyle='solid')

plt.xlim([-0.04,1.04])
plt.ylim([-0.04, 1.04])

textcolor = 'k'

plt.xlabel(r"$t$", fontsize=26, color = textcolor)
plt.ylabel(r"$x(t)$", fontsize=26, color=textcolor)

plt.tick_params(axis='x', which='both', top='off', color = textcolor)
plt.xticks(fontsize=24, rotation=0, color = textcolor)
plt.tick_params(axis='y', which='both', right='off', color = textcolor)
plt.yticks(fontsize=24, rotation=0, color = textcolor)
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=6)

plt.tight_layout()

fig.set_size_inches(8, 6)

plt.savefig('lev_trajectory' + '.jpg', dpi=600)
plt.show()
```

```python
fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.rcParams['figure.dpi']= 600

ax.grid()
ax.set_axisbelow(True)

t = np.linspace(0,1, 600)
x = (4./3.)*np.cbrt(t-1.)
plt.plot(t,x, color='xkcd:dark sky blue', linewidth='2.5', linestyle='solid')

plt.xlim([-0.04,1.04])
plt.ylim([-1.35, 0])

textcolor = 'k'

plt.xlabel(r"$t$", fontsize=26, color = textcolor)
plt.ylabel(r"$x'(t)$", fontsize=26, color=textcolor)

plt.tick_params(axis='x', which='both', top='off', color = textcolor)
plt.xticks(fontsize=24, rotation=0, color = textcolor)
plt.tick_params(axis='y', which='both', right='off', color = textcolor)
plt.yticks(fontsize=24, rotation=0, color = textcolor)
ax.locator_params(axis='y', nbins=6)
ax.locator_params(axis='x', nbins=6)


plt.tight_layout()

fig.set_size_inches(8, 6)

plt.savefig('lev_velocity' + '.jpg', dpi=600)
plt.show()
```
